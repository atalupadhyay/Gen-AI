# Understanding Statistical Distributions in Python: A Comprehensive Guide

**Introduction**

Statistical distributions are the cornerstone of data analysis and machine learning. They describe how data is spread across a range of values, providing essential insights into the underlying patterns and probabilities. In this post, we'll explore some of the most common statistical distributions, how they are used, and how to work with them in Python.

**Why are Statistical Distributions Important?**

Statistical distributions help us to:

*   **Understand Data:** Visualize and interpret the characteristics of a dataset (e.g., mean, variance, skewness).
*   **Make Predictions:** Estimate probabilities and predict future outcomes based on observed data.
*   **Model Phenomena:** Create mathematical models that represent real-world processes.
*   **Perform Statistical Tests:** Determine if observed differences in data are statistically significant.
*   **Generate Realistic Data:** Sample from distributions to simulate realistic data for testing and modeling.

**Essential Libraries in Python**

Before we begin, let's introduce the core Python libraries we'll be using:

*   **NumPy:** For numerical operations and generating random data.
*   **SciPy:** For statistical distributions, functions, and analysis tools.
*   **Matplotlib and Seaborn:** For visualizing distributions.

**Setting up the Environment**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Configure plot settings
sns.set_theme()  # For better plot styling
```

**Common Statistical Distributions**

Let's dive into some of the most commonly encountered distributions, discussing their properties, uses, and implementation in Python.

**1. Normal (Gaussian) Distribution**

The normal distribution, often referred to as the "bell curve," is one of the most important distributions in statistics.

*   **Properties:**
    *   Symmetrical around the mean (average).
    *   Mean, median, and mode are all equal.
    *   Defined by two parameters: the mean (μ) and standard deviation (σ).
    *   68% of the data falls within 1 standard deviation of the mean, 95% within 2 standard deviations, and 99.7% within 3 standard deviations.

*   **Uses:**
    *   Many natural phenomena are approximately normally distributed (e.g., heights, weights, test scores).
    *   Forms the basis for many statistical tests and modeling techniques.

*   **Python Implementation:**

```python
# Parameters
mu = 0  # Mean
sigma = 1 # Standard deviation

# Generate samples
normal_data = np.random.normal(mu, sigma, 1000)

# Plot the Distribution
sns.histplot(normal_data, kde=True, label="Normal Distribution")
plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# Probability density function (PDF)
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 200)
pdf = stats.norm.pdf(x, mu, sigma)
plt.plot(x, pdf, label="PDF")
plt.title("Normal Distribution PDF")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.show()

# Cumulative distribution function (CDF)
cdf = stats.norm.cdf(x, mu, sigma)
plt.plot(x, cdf, label="CDF")
plt.title("Normal Distribution CDF")
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.legend()
plt.show()
```

**2. Uniform Distribution**

The uniform distribution assigns equal probability to all values within a given range.

*   **Properties:**
    *   All values within a range have the same probability.
    *   Defined by two parameters: the minimum (a) and maximum (b) values of the range.

*   **Uses:**
    *   Modeling random processes where each outcome is equally likely (e.g., rolling a fair die).
    *   Used in Monte Carlo simulations and random number generation.

*   **Python Implementation:**

```python
# Parameters
a = 0 # Minimum value
b = 1 # Maximum value

# Generate samples
uniform_data = np.random.uniform(a, b, 1000)

# Plot the Distribution
sns.histplot(uniform_data, kde=True, label="Uniform Distribution")
plt.title("Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# Probability density function (PDF)
x = np.linspace(a - 0.5, b + 0.5, 200)
pdf = stats.uniform.pdf(x, a, b-a)
plt.plot(x, pdf, label="PDF")
plt.title("Uniform Distribution PDF")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.show()

# Cumulative distribution function (CDF)
cdf = stats.uniform.cdf(x, a, b-a)
plt.plot(x, cdf, label="CDF")
plt.title("Uniform Distribution CDF")
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.legend()
plt.show()
```

**3. Binomial Distribution**

The binomial distribution describes the probability of a number of successes in a fixed number of independent trials.

*   **Properties:**
    *   Discrete distribution.
    *   Defined by two parameters: the number of trials (n) and the probability of success in each trial (p).

*   **Uses:**
    *   Modeling the probability of a specific number of successful events (e.g., the probability of flipping 3 heads in 5 coin tosses).
    *   Used in hypothesis testing and quality control.

*   **Python Implementation:**

```python
# Parameters
n = 10  # Number of trials
p = 0.5 # Probability of success

# Generate samples
binomial_data = np.random.binomial(n, p, 1000)

# Plot the Distribution
sns.histplot(binomial_data, kde=False, label="Binomial Distribution")
plt.title("Binomial Distribution")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# Probability mass function (PMF)
x = np.arange(0, n + 1)
pmf = stats.binom.pmf(x, n, p)
plt.plot(x, pmf, 'o-', label="PMF")
plt.title("Binomial Distribution PMF")
plt.xlabel("Number of Successes")
plt.ylabel("Probability Mass")
plt.legend()
plt.show()

# Cumulative distribution function (CDF)
cdf = stats.binom.cdf(x, n, p)
plt.plot(x, cdf, label="CDF")
plt.title("Binomial Distribution CDF")
plt.xlabel("Number of Successes")
plt.ylabel("Cumulative Probability")
plt.legend()
plt.show()
```

**4. Poisson Distribution**

The Poisson distribution models the probability of a certain number of events occurring within a fixed interval of time or space.

*   **Properties:**
    *   Discrete distribution.
    *   Defined by a single parameter: the average rate of events (λ).

*   **Uses:**
    *   Modeling the number of phone calls per hour, the number of cars passing a certain point on a road, or the number of defects in a product.
    *   Used in queuing theory, traffic analysis, and risk management.

*   **Python Implementation:**

```python
# Parameters
lam = 5  # Average rate of events

# Generate samples
poisson_data = np.random.poisson(lam, 1000)

# Plot the Distribution
sns.histplot(poisson_data, kde=False, label="Poisson Distribution")
plt.title("Poisson Distribution")
plt.xlabel("Number of Events")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# Probability mass function (PMF)
x = np.arange(0, 20)
pmf = stats.poisson.pmf(x, lam)
plt.plot(x, pmf, 'o-', label="PMF")
plt.title("Poisson Distribution PMF")
plt.xlabel("Number of Events")
plt.ylabel("Probability Mass")
plt.legend()
plt.show()

# Cumulative distribution function (CDF)
cdf = stats.poisson.cdf(x, lam)
plt.plot(x, cdf, label="CDF")
plt.title("Poisson Distribution CDF")
plt.xlabel("Number of Events")
plt.ylabel("Cumulative Probability")
plt.legend()
plt.show()
```

**5. Exponential Distribution**

The exponential distribution models the time between events in a Poisson process.

*   **Properties:**
    *   Continuous distribution.
    *   Defined by a single parameter: the rate parameter (λ).

*   **Uses:**
    *   Modeling the lifespan of devices or components, the time between customer arrivals, or the time between earthquakes.

*   **Python Implementation:**

```python
# Parameters
lam = 0.5 # Rate parameter

# Generate samples
exponential_data = np.random.exponential(1/lam, 1000) #Note that we need to pass 1/lambda to the function

# Plot the Distribution
sns.histplot(exponential_data, kde=True, label="Exponential Distribution")
plt.title("Exponential Distribution")
plt.xlabel("Time Between Events")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# Probability density function (PDF)
x = np.linspace(0, 10, 200)
pdf = stats.expon.pdf(x, scale = 1/lam)
plt.plot(x, pdf, label="PDF")
plt.title("Exponential Distribution PDF")
plt.xlabel("Time Between Events")
plt.ylabel("Probability Density")
plt.legend()
plt.show()

# Cumulative distribution function (CDF)
cdf = stats.expon.cdf(x, scale = 1/lam)
plt.plot(x, cdf, label="CDF")
plt.title("Exponential Distribution CDF")
plt.xlabel("Time Between Events")
plt.ylabel("Cumulative Probability")
plt.legend()
plt.show()
```
